```python
import pandas as pd
from langchain_community.document_loaders import DataFrameLoader

g_df = pd.read_csv('ground_truth_1.csv')

g_loader = DataFrameLoader(g_df, page_content_column='review')
g_docs = g_loader.load()

from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("""
    주어진 제품명과 문맥에 따라, 그 제품에 대한 구체적인 답변을 작성해 주세요.
    
    제품명: {goodsName}
    문맥: {contexts}

    질문에 대한 간결하고 정확한 답변을 생성하세요.
    """)
    
model = ChatOpenAI(model='gpt-4o', temperature=0)

output_parser = StrOutputParser()

ground_truth_chain = prompt | model | output_parser

# 답변 생성
num = 0
ground_truth_data1 = []
for doc in g_docs:
    
    doc_context = g_df['review'].iloc[num]  # Assuming 'context' is a column in df
    doc_goodsName = doc.metadata['goodsName']
    
    # Invoke the chain with goodsName and context
    result = ground_truth_chain.invoke({'goodsName': doc_goodsName, 'contexts': doc_context})
    
    # Append the result to the ground_truth_data list
    ground_truth_data1.append(result)
    num += 1

prompt_eval_df['ground_truth'] = ""  # 새로운 컬럼 생성
prompt_eval_df.loc[0, 'ground_truth'] = str(ground_truth_data1)
```

```python
from langchain.chat_models import ChatOpenAI
import pandas as pd
import chardet

def detect_encoding(file_path):
    """파일의 인코딩을 자동으로 감지"""
    with open(file_path, 'rb') as f:
        raw_data = f.read()
    result = chardet.detect(raw_data)
    return result['encoding']

def generate_prompt_from_csv(csv_file_path, template_file_path, role='ai'):
    """CSV 파일을 읽어 각 행에 대해 프롬프트를 생성"""
    # CSV 파일의 인코딩을 감지하여 읽음
    encoding = detect_encoding(csv_file_path)
    df = pd.read_csv(csv_file_path, encoding=encoding)

    # 템플릿 파일을 불러오기
    encoding = detect_encoding(template_file_path)  # 템플릿 파일 인코딩 감지
    with open(template_file_path, 'r', encoding=encoding) as template_file:
        template_content = template_file.read()

    # 각 행을 기반으로 프롬프트 생성
    prompts = []
    for _, row in df.iterrows():
        # 행을 딕셔너리로 변환
        row_data = row.to_dict()
        row_data['role'] = role  # role 추가
        
        # 템플릿 포맷팅
        prompt = template_content.format(**row_data)
        prompts.append(prompt)
    
    return prompts

# 평가 프롬프트 생성 실행
evaluation_prompts = generate_prompt_from_csv('rag_test_dataset_sofar.csv', 'prompt_template_ai.yaml')

critics_model = ChatOpenAI(model='gpt-4o', temperature=0)

# 각 프롬프트에 대해 모델을 3번 호출하여 평가
evaluations_final = []
for prompt in evaluation_prompts:
    evaluations = []
    for _ in range(3):  # 3번 평가
        response = critics_model.invoke(prompt)  # 포맷된 프롬프트를 모델에 전달
        evaluations.append(response.content)
    
    evaluations_final.append(evaluations)
```
